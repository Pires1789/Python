## Comandos b√°sicos do Pandas

```python
#Criando um dataframe

data = {'Nome': ['Alice', 'Bob', 'Charlie'],
        'Idade': [25, 30, 35],
        'Cidade': ['Nova York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
```

```python
#Primeiras cinco colunas do dataframe.

df.head()
```

```python
#Fun√ß√£o para apresentar uma lista com o t√≠tulo das colunas.
df.columns
```

```python
#Forma de apresentar apenas uma coluna de um DataFrame

df[['Cidade']]
```

```python
#Selecionando mais de uma coluna

df[['Cidade', 'Nome']]
df
```

```python
#Selecionando colunas com base em condi√ß√µes

jovens = df[df['Idade'] < 30]
jovens
```

```python
#Adicionar colunas

df['Sal√°rio'] = [50000, 60000, 75000]
df
```

```python
#Realizando agrega√ß√µes

media_idade = df['Idade'].mean()
soma_salarios = df['Sal√°rio'].sum()

media_idade
```
