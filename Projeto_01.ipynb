# Projeto 01: pacientes com diabetes

```python
# importando as bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
#Importando o bando de dados

df = pd.read_csv('/content/drive/MyDrive/03 - BI & Ciência de Dados/Programação/Python/Lab01 - DSA Análise de Dados com Linguagem Python/diabetes.csv')
df.head()
```

```python
# Função para validar o IMC (BMI) de cada linha e retornar se o paciente está dentro do peso ou fora
def classificar_bmi(bmi):
    if bmi < 30:
        return 'Peso ideal'
    else:
        return 'Obeso'

#Executando a função
df['Classificação'] = df['BMI'].apply(classificar_bmi)

df.head()
```

```python
#Criando dataframes para análise de dados
df_50_anos = df[(df['Age'] > 50) & (df['Outcome'] == 1)]
df_50_anos_com_diabetes = df_50_anos[df_50_anos['Outcome']==1]
df_50_anos_sem_diabetes = df_50_anos[df_50_anos['Outcome']==0]

df_50_anos.head()
```

# Explorando Dados

```python
#Comparação entre "Obsesos" e "Peso Ideal"

contagem_classificacao = df_50_anos['Classificação'].value_counts()

fig = go.Figure(data=go.Pie(labels=contagem_classificacao.index,
                            values=contagem_classificacao.values))

fig.update_layout(title='Percentual de Obesos e Peso Ideal',
                  showlegend=True)

fig.show()
```

```python
#Avaliando o percentual de obesos e peso ideal

obseso = (df['BMI']>30).sum()
magro = (df['BMI']<30).sum()
valoes = [obseso, magro]
labels = ['obseso', 'magro']

fig = go.Figure(data=go.Pie(labels=contagem_classificacao.index,
                            values=contagem_classificacao.values))

fig.update_layout(title='Percentual de Obesos e Peso Ideal',
                  showlegend=True)

fig.show()
```

```python
maior_50 = (df['Age'] > 50).sum()
menor_50 = (df['Age'] < 50).sum()
valores = [maior_50, menor_50]
rotulos = ['maior_50', 'menor_50']


# Criando o gráfico de barras
fig_barra = go.Figure(data=go.Bar(x=rotulos, y=valores))

# Personalizando layout
fig_barra.update_layout(title='Gráfico de Barras',
                  xaxis_title='Categoria',
                  yaxis_title='Valores')

fig_barra.show()
```
